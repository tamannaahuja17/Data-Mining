```python
import pandas as pd
import numpy as np
```

```python
data={'Age': np.random.randint( 30,70,100),'Income': np.random.randint(20000,100000,100)}
df=pd.DataFrame(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
std_data = pd.DataFrame(scaler.fit_transform(df),columns=['Age','Income'])
display(std_data)
std_data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
normalized_data = pd.DataFrame(scaler.fit_transform(df),columns = ['Age','Income'])
display(normalized_data)
```

```python
df_transformed = df.copy()
df_transformed['Income'] = np.log(df_transformed['Income'])
df_transformed
```

```python
df_aggregated= df.copy()
df_aggregated = df.groupby('Age').mean().reset_index()
df_aggregated.head()
```

```python
bins = [19,30,50,np.inf]
labels = ['Young','Adult','Elderly']

df_descretized = df.copy()
df_descretized['Age group'] = pd.cut(df_descretized['Age'],bins = bins, labels = labels)
df_descretized.head()
```

```python
from sklearn.preprocessing import Binarizer
threshold = 30
binarizer = Binarizer(threshold = threshold)
df_binarized = df.copy()
df_binarized = binarizer.fit_transform(df[['Age']])
df_binarized= pd.DataFrame(df_binarized, columns=['Binarized_Age'])
df_binarized=pd.concat([df,df_binarized],axis=1)
df_binarized.head()
```

```python
sampled_data = df.sample(frac=0.1,random_state = 1)
sampled_data
```
