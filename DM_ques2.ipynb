```python
import pandas as pd
import numpy as np
```

```python
data={'Age': np.random.randint( 30,70,100),'Income': np.random.randint(20000,100000,100)}
df=pd.DataFrame(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
std_df = pd.DataFrame(scaler.fit_transform(df),columns=['Age','Income'])
display(std_df)
std_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
norm_df = pd.DataFrame(scaler.fit_transform(df),columns = ['Age','Income'])
display(norm_df)
```

```python
transformed_df = df.copy()
transformed_df['Income'] = np.log(df['Income'])
transformed_df
```

```python
aggregated_df = df.groupby('Age').mean()
aggregated_df.head()
```

```python
bins = [19,30,50,100]
labels = ['Young','Adult','Elderly']

descretized_df = df.copy()
descretized_df['Age group'] = pd.cut(descretized_df['Age'],bins = bins, labels = labels)
descretized_df.head()
```

```python
from sklearn.preprocessing import Binarizer
threshold = 30 #Age < 30 will be 0 and > = 30 will be 1
binarizer = Binarizer(threshold = threshold)
binarized_df = df.copy()
binarized_df['Binarized_age'] = binarizer.fit_transform(df[['Age']])
binarized_df
```

```python
sampled_data = df.sample(frac=0.1,random_state = 1)
sampled_data
```
