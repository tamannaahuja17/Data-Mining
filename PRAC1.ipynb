```python
rule1=data_iris['Species'].apply(lambda x:True if x=='setosa' or x=='versicolor' else False)
rule1
```

```python
rule2=data_iris['Petal.Length'].apply(lambda x: True if x==2*'petal width'else False)
rule2
```

```python
Rules = pd.DataFrame({"Rule1":rule1,"Rule2":rule2})
Rules
```

```python
Rules.astype(int)
```

```python
print("number of rules violated : ",len(Rules)-Rules["Rule1"].sum())
```

```python
print("count of both\n",Rules["Rule1"].value_counts())
```

```python
print("number of rules violated : ",len(Rules)-Rules["Rule2"].sum())
```

```python
print("count of both\n",Rules["Rule2"].value_counts())
```

```python
import matplotlib.pyplot as plt
plt.figure()
Rules.apply(lambda x:len(x)-x.sum()).plot(kind='bar')
plt.xlabel="Rules"
plt.ylabel="Number of records that violates the rules"
```
