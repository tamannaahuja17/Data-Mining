```python
import pandas as pd
from efficient_apriori import apriori
df = pd.read_csv("groceriesDataset.csv")
df
```

```python
df.fillna(0,inplace=True)
df
```

```python
transactions = []
for index, row in df.iterrows():
    transaction = [str(item) for item in row if pd.notna(item) and item!=0 and str(item).isnumeric()==False]
    transactions.append(transaction)
transactions
```

```python
itemsets_1, rules_1 = apriori(transactions, min_support=0.04, min_confidence=0.15)
itemsets_1
```

```python
rules_1
```

```python
itemsets_2, rules_2 = apriori(transactions, min_support=0.06, min_confidence=0.06)
itemsets_2
```

```python
rules_2
```
